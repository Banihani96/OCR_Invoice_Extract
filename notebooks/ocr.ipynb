```python
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel , Field
from typing import List, Optional
class InvoiceItem(BaseModel):
    description : str = Field(..., description="item's description")
    taxes : Optional[float] = Field(..., description="item's taxes")

class InvoiceDate(BaseModel):
    buyer_name: str = Field(..., description="buyer name in the invoice")
    invoice_date: str = Field(..., description="invoice date")
    items: List[InvoiceItem] = Field(..., description="Items data in the invoice")

parser = PydanticOutputParser(pydantic_object = InvoiceDate)
```
